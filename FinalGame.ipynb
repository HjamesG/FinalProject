```python
import random
import time

rooms = [["###   ###",
          "###   ###",
          "###   ###",
          "###   ###",
          "###   ###",
          "###   ###",
          "###   ###",
          "###   ###",
          "###   ###", "UD"],
         
         ["#########",
          "#########",
          "#########",
          "         ",
          "         ",
          "         ",
          "#########",
          "#########",
          "#########", "LR"],
         
         ["###   ###",
          "###   ###",
          "###   ###",
          "###      ",
          "###      ",
          "###      ",
          "#########",
          "#########",
          "#########", "UR"],  
         
         ["###   ###",
          "###   ###",
          "###   ###",
          "      ###",
          "      ###",
          "      ###",
          "#########",
          "#########",
          "#########", "UL"],
         
         ["#########",
          "#########",
          "#########",
          "###      ",
          "###      ",
          "###      ",
          "###   ###",
          "###   ###",
          "###   ###", "RD"],
         
         ["#########",
          "#########",
          "#########",
          "      ###",
          "      ###",
          "      ###",
          "###   ###",
          "###   ###",
          "###   ###", "UD"],
         
         ["###   ###",
          "###   ###",
          "###   ###",
          "         ",
          "       O ",
          "         ",
          "###   ###",
          "###   ###",
          "###   ###", "URDL"],
         
         ["###   ###",
          "###   ###",
          "###   ###",
          "###      ",
          "###      ",
          "###      ",
          "###   ###",
          "###   ###",
          "###   ###", "URD"],
         
         ["###   ###",
          "###   ###",
          "###   ###",
          "      ###",
          "      ###",
          "      ###",
          "###   ###",
          "###   ###",
          "###   ###", "UDL"],
         
         ["###   ###",
          "###   ###",
          "###   ###",
          "         ",
          "         ",
          "         ",
          "#########",
          "#########",
          "#########", "URL"],
         
         ["###   ###",
          "###   ###",
          "##     ##",
          "         ",
          "    O    ",
          "         ",
          "##     ##",
          "###   ###",
          "###   ###", "URDL"],
         
         ["#########",
          "#########",
          "#########",
          "         ",
          "         ",
          "         ",
          "###   ###",
          "###   ###",
          "###   ###", "RDL"],
        ]

player = [0, 0, "U", 4, 4]

start = int(time.time())

playTime = 45

world = [[rooms[6]]]

# [1, 2, 3],
# [4, 5, 6],
# [7, 8, 9]

timer = 1

def printScreenAndPlayer():
    oldScreen = [world[player[0]][player[1]]]
    newScreen = []
    for i in range(0,9):
        newScreen.append(list(oldScreen[0][i]))
         
    newScreen[player[3]][player[4]] = "P"
    for r in range(0,9):
        line = ""
        
        for c in range(0,9):
            line = line + newScreen[r][c] + " "
            
        if r == 3:    
            timer = playTime - round(time.time() - start, 1)
            line = line + "       Time Remaining: " + str(timer)
        if r == 4:
            line = line + "       Time is running!"
        print(line)
    newScreen[player[3]][player[4]] = " "
    
# Test case:
# printScreenAndPlayer()

def expandUp():
    temp = []
    for i in range(0, len(world[0])):
        temp.append("")
    world.insert(0, temp)
    player[0] = player[0] + 1
                    
def expandDown():
    temp = []
    for i in range(0, len(world[0])):
        temp.append("")
    world.append(temp)

def expandRight():
    for i in range(0, len(world)):
        world[i].append("")

def expandLeft():
    for i in range(0, len(world)):
        world[i].insert(0,"")
    player[1] = player[1] + 1
    
def moveUp():
    if player[3] == 0:
        if player[0] == 0:
            expandUp()

        
        player[0] = player[0] - 1
        player[3] = 8
        # make new room
        plausable = []
        for x in range(0, len(rooms)):
            for l in range(0,len(rooms[x][-1])):
                if rooms[x][-1][l] == "D":
                    plausable.append(rooms[x])
        world[player[0]][player[1]] = plausable[random.randint(0, len(plausable)-1)]

        #print(world)
    elif world[player[0]][player[1]][player[3]-1][player[4]] == "#":
        return
    
    elif world[player[0]][player[1]][player[3]-1][player[4]] == "O":
        start = time.time()
        player[3] = player[3] - 1

    else:
        player[3] = player[3] - 1
        
def moveDown():
    if player[3] == 8:
        if player[0] == len(world)-1:
            expandDown()
        #if world[player[0]+1][player[1]] == "":
        player[0] = player[0] + 1
        player[3] = 0
        # make new room
        plausable = []
        for x in range(0, len(rooms)):
            for l in range(0,len(rooms[x][-1])):
                if rooms[x][-1][l] == "U":
                    plausable.append(rooms[x])
        world[player[0]][player[1]] = plausable[random.randint(0, len(plausable)-1)]

    elif world[player[0]][player[1]][player[3]+1][player[4]] == "#":
        return
    
    elif world[player[0]][player[1]][player[3]+1][player[4]] == "O":
        start = time.time()
        player[3] = player[3] + 1
        
    else:
        player[3] = player[3] + 1
        
def moveLeft():
    if player[4] == 0:
        if player[1] == 0:
            expandLeft()
        #if world[player[0]][player[1]-1] == "":
        player[1] = player[1] - 1
        player[4] = 8
        # make new room
        plausable = []
        for x in range(0, len(rooms)):
            for l in range(0,len(rooms[x][-1])):
                if rooms[x][-1][l] == "R":
                    plausable.append(rooms[x])
        world[player[0]][player[1]] = plausable[random.randint(0, len(plausable)-1)]

        #print(world)
    elif world[player[0]][player[1]][player[3]][player[4]-1] == "#":
        return
    
    elif world[player[0]][player[1]][player[3]][player[4]-1] == "O":
        start = time.time()
        player[4] = player[4] - 1

    else:
        player[4] = player[4] - 1
    
def moveRight():
    if player[4] == 8:
        if player[1] == len(world[0])-1:
            expandRight()
        #if world[player[0]][player[1]+1] == "":
        player[1] = player[1] + 1
        player[4] = 0
        # make new room
        plausable = []
        for x in range(0, len(rooms)):
            for l in range(0,len(rooms[x][-1])):
                if rooms[x][-1][l] == "L":
                    plausable.append(rooms[x])
        world[player[0]][player[1]] = plausable[random.randint(0, len(plausable)-1)]

    elif world[player[0]][player[1]][player[3]][player[4]+1] == "#":
        return
    
    elif world[player[0]][player[1]][player[3]][player[4]+1] == "O":
        start = time.time()
        player[4] = player[4] + 1

    else:
        player[4] = player[4] + 1
    


def gamePlay():    
    while True:
        if playTime <= (time.time() - start):
            print("\n\n GAME OVER. YOU DIED.\n\n")
            return
        printScreenAndPlayer()
        usr = input("")
        if usr == "w":
            moveUp()
        elif usr == "s":
            moveDown()
        elif usr == "a":
            moveLeft()
        elif usr == "d":
            moveRight()

gamePlay()
```
